### PULLING DATA FROM WIKIPEDIA
#### LIST OF COMPANIES IN NIGERIA

https://en.wikipedia.org/wiki/List_of_companies_of_Nigeria

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_Nigeria'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
cmp = soup.find_all('th')[:6]
cmp
```

```python
cmp = [title.text.strip() for title in cmp]
print(cmp)
```

```python
df = pd.DataFrame(columns = cmp)
df
```

```python
cln = soup.find_all('tr')
```

```python
for row in cln[2:34]:
    row = row.find_all('td')
    eachRow = [data.text.strip() for data in row]
    
    lth = len(df)
    df.loc[lth] = eachRow
```

```python
df.to_csv(r'C:\Users\Mhizta Joe\Desktop\MrJoe\list.csv')
df
```
