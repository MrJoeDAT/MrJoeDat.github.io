<h2>Rename EmployeeNumber Column

<h2>Move EmployeeID to First Column

```python
#Move EmployeeID to first column
mov = 'EmployeeID'
new_position = 0

#Create copy of the DataFrame with column rearranged
columns = df.columns.tolist()
columns.remove(mov)
columns.insert(new_position, mov)

#Reorder DataFrame columns
df = df[columns]

df
```

<span style = "color:green; font-style:italic">Made a little changes to the arrangement by moving EmployeeID to first column</span>

<h2>Correlation Map for Numeric Variables

```python
#Correlation map for numeric variables
corr_data = ['DailyRate', 'MonthlyIncome','MonthlyRate', 'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
corr_sample = df[corr_data].copy()

correlation_matrix = corr_sample.corr()
plt.figure(figsize = (12, 8))
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
```

<span style = "color:green; font-style:italic">Plotted a correlation map for the numeric variables  
    It shows the relationship between the variables</span>

<h2>Distribution of OverTime by EmployeeID

```python
#Distribution of OverTime by EmployeeID
plt.figure(figsize = (8, 6))
value_counts = df['OverTime'].value_counts()
ax = value_counts.plot(kind = 'bar', color = ['royalblue', 'orange'])
ax.set_xticks([0, 1]) #Set tick position
ax.set_xticklabels(['No', 'Yes']) #Set custom tick labels
plt.title('Distribution of OverTime by EmployeeID')
plt.xlabel('EmployeeID')
plt.ylabel('Count of EmployeeID')
plt.xticks(rotation = 0)
plt.show()
```

<span style = "color:green; font-style:italic">Bar chart that shows distribution of OverTime by Employees  
We can see that most employee don't work overtime</span>

# 

<h2>Donut Chart to Show Marital Status

```python
#Donut chart to show Marital Status
marital_status_counts = df['MaritalStatus'].value_counts()
fig, ax = plt.subplots()
ax.pie(marital_status_counts, labels = marital_status_counts.index, autopct = '%1.1f%%', startangle = 90, pctdistance = 0.85)

#Draw a circle in the center to make it a donut chart
circle = plt.Circle((0, 0), 0.70, fc = 'white')
fig.gca().add_artist(circle)

#Equal aspect ratio ensures that pie is drawn as a circle
ax.axis('equal')

plt.title('Marital Status Distribution')
plt.show()
```

<span style = "color:green; font-style:italic">This donut chart shows that almost half of the employees are married and there are more employees that are single than divorced</span>

<h2>Chart of Number of Companies Worked

```python
#Create bins for 'NumCompaniesWorked'
bins = [0, 2, 4, 6, 7] #Bin edges

#Create a new column with categories based on the bins
df['NumCompaniesCategory'] = pd.cut(df['NumCompaniesWorked'], bins = bins, labels = ['0 - 2', '3 - 4', '5 - 6', '7 + '])

#Count the occurrences of each category for each group (in this case, it's the entire dataset)
category_counts = df['NumCompaniesCategory'].value_counts().sort_index()

#Create a stacked bar chart
plt. figure(figsize = (8, 6))
category_counts.plot(kind = 'barh', stacked = True, color = 'royalblue')
plt.title('Stacked Bar Chart of Number of Companies Worked Categories')
plt.xlabel('Count')
plt.ylabel('Categories')
plt.show()
```

<span style = "color:green; font-style:italic">Chart of numbers of companies worked by employees  
Huge number of employee fall under 0 - 2 and there are some employee that have worked with more than 7 companies</span>
