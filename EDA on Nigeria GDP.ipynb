# EXPLORATORY DATA ANALYSIS ON NIGERIA GDP 

### Using Nigeria GDP Dataset

https://www.cbn.gov.ng/rates/RealGDP.asp

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#The dataset values are in BILLION and NAIRA
df = pd.read_csv(r'C:\Users\Mhizta Joe\Desktop\MrJoe\Nig\gdp.csv')
df
```

```python
#Get rid of datas from 1981 to 2009
df2 = df.drop(range(0, 29)).copy() 
df2
```

```python
#Delete the last row because it's the current year(not complete)
df2 = df2.drop(65)
df2
```

```python
#Reset the default index
df2.reset_index(drop = True, inplace = True)
df2
```

```python
#Brief description of the dataset
df2.describe()

#We can see the year with minimum and maximum GDP
#We can also see the average GDP of each sector(mean)
```

```python
#We'll focus on 3 major sectors that affect the overall GDP of the country
df3 = df2[['Year', 'Period', 'Agriculture', 'Industry', 'CrudePetroleumAndNaturalGas']]
df3
```

```python
#Visualizing Annual values from the dataset
Annual = df3[df3['Period'].str.contains('Annual')]
Annual
```

```python
#Visualizing Quarters values from the dataset
Quarter = df3[df3['Period'].str.contains('Q')]
Quarter
```

```python
#The average of each quarters in Agriculture
Quarter.groupby('Period')['Agriculture'].mean()
```

```python
#Correlation of the three sectors
df3.corr()
```

```python
#Heatmap that shows the correlation
sns.heatmap(Quarter.corr(), annot = True)
plt.rcParams['figure.figsize'] = (20,10)
plt.show()
```

```python
#Annual performance of Agriculture from 2010 to 2022
Annual[['Year', 'Agriculture']].sort_values(by='Agriculture', ascending=False, ignore_index = True)


#This sector experienced steady growth regardless of the global pandemic that started from 2019
```

```python
#Annual performance of Oil from 2010 to 2022
Annual[['Year', 'CrudePetroleumAndNaturalGas']].sort_values(by='CrudePetroleumAndNaturalGas', ascending=False, ignore_index = True)

#This sector decline in growth from 2019 because of the pandemic(movements were restricted for a while)
```
