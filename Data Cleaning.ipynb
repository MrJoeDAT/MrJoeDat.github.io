# DATA CLEANING WITH PYTHON

### Operations to be performed

#### Removing unwanted characters, Creating new column from old ones, Removing duplicates, Replacing characters, etc

# 

```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Mhizta Joe\Desktop\MrJoe\listEdited.csv', index_col = 0)
df
```

```python
df["Headquarters"] = df["Headquarters"].str.strip("/_.")
df
```

```python
df['Founded'] = df['Founded'].str.replace('[^a-zA-Z0-9]', '')
df
```

```python
df[['Industry', 'Sector']] = df['Industry-Sector'].str.split(',', 2, expand = True)
df
```

```python
df['Ignore'] = df['Ignore'].str.replace('yes', 'Y')
df['Ignore'] = df['Ignore'].str.replace('no', 'N')
df['Ignore'] = df['Ignore'].str.replace('y', 'Y')
df['Ignore'] = df['Ignore'].str.replace('n', 'N')
df
```

```python
df = df.drop(columns = ['Industry-Sector', 'Ignore'])
df
```

```python
df = df.drop_duplicates()
df
```
