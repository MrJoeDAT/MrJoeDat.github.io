## ANALYSING THE IMPACT OF THE COVID-19 PANDEMIC ON THE NIGERIAN ECONOMY

### Using Nigeria GDP and Inflation dataset

   GDP - https://www.cbn.gov.ng/rates/RealGDP.asp

   Inflation - https://www.cbn.gov.ng/rates/inflrates.asp

### We'll start with pulling the datasets of Nigeria GDP

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
gdp = pd.read_csv(r'C:\Users\Mhizta Joe\Desktop\MrJoe\gdpann.csv')
ir = pd.read_csv(r'C:\Users\Mhizta Joe\Desktop\MrJoe\irann.csv')
```

## The GDP Dataset

```python
gdp.head(10)
```

```python
yr = gdp['Year']
agr = gdp['Agriculture']
ind = gdp['Industry']
oil = gdp['CrudePetroleumAndNaturalGas']
re = gdp['RealEstate']

plt.figure(figsize=(20, 10))

plt.plot(yr, agr, label = 'Agriculture')
plt.plot(yr, ind, label = 'Industry')
plt.plot(yr, oil, label = 'CrudePetroleumAndNaturalGas')
plt.plot(yr, re, label = 'RealEstate')

plt.xlabel('Year')
plt.ylabel('Sectors')
plt.title('GDP Flow from 2010 to 2022')
plt.legend(loc='upper left')

plt.show()
```

### In the line chart above, we can study how Nigeria GDP have been moving from 2010 to 2022 in each sector
### I discovered that agriculture experienced steady growth regardless of the pandemic
### The industry sector experienced steady growth from 2016 and started declining from 2019, same with oil sector

## The Inflation Dataset

```python
ir.head(10)
```

```python
yr = ir['Year']
aitChn = ir['All Items (Year on Change)']
aitAvgChn = ir['All Items (12 Months Avg. Change)']
foodChn = ir['Food (Year on Change)/1']
foodAvgChn = ir['Food (12 Months Avg. Change)/1']

plt.figure(figsize=(20, 10))

plt.bar(yr, aitChn, bottom = 0, color = 'blue', label = 'All Items (Year on Change)')
plt.bar(yr, aitAvgChn, bottom = aitAvgChn + foodChn, color = 'lightblue', label = 'All Items (12 Months Avg. Change)')
plt.bar(yr, foodChn, bottom = aitChn, color = 'gold', label = 'Food (Year on Change)/1')
plt.bar(yr, foodAvgChn, bottom = aitAvgChn + foodChn + foodAvgChn, color = 'yellow', label = 'Food (12 Months Avg. Change)/1')

plt.xlabel('Year')
plt.ylabel('Sectors')
plt.title('Inflation Rate from 2010 to 2022')
plt.legend(loc='upper left')

plt.show()
```

### In the bar chart above all highlighted sectors experienced increase right from 2019 when the pandemic hit
